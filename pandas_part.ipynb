```python
"""
This script reads and prints the contents of a CSV file named 'BTC-USD.csv' using pandas.

Steps:
1. Imports the pandas library.
2. Reads the CSV file 'BTC-USD.csv' into a DataFrame called `data`.
3. Prints the entire DataFrame to the console.

This is a basic example of loading and displaying tabular data using pandas.
"""

import pandas as pd
data = pd.read_csv("BTC-USD.csv")
print(data)
```

```python
"""
This command displays a concise summary of the DataFrame `data`.

It shows:
- The number of entries (rows).
- The total number of columns.
- Column names and their data types.
- Number of non-null (non-missing) values in each column.
- Memory usage of the DataFrame.

This is useful for quickly understanding the structure and completeness of the dataset.
"""

data.info()
```

```python
"""
This command generates descriptive statistics of the numerical columns in the DataFrame `data`.

It includes:
- Count (number of non-null values)
- Mean (average)
- Standard deviation (std)
- Minimum value (min)
- 25th percentile (25%)
- Median or 50th percentile (50%)
- 75th percentile (75%)
- Maximum value (max)

This summary helps understand the distribution and spread of the dataset's numerical features.
"""

data.describe()
```

```python
"""
This script modifies the 'Open' column of the DataFrame `data` by multiplying each value by 10.

Steps:
1. Accesses the 'Open' column of the DataFrame.
2. Multiplies every value in this column by 10.
3. Updates the 'Open' column with the new values.
4. Prints the updated DataFrame to show the changes.

This demonstrates how to perform element-wise operations on a specific DataFrame column.
"""

data['Open'] = data['Open'] * 10
print(data)
```

```python
"""
This script visualizes a segment of the 'Open' and 'Close' prices from the DataFrame `data` using line plots.

Steps:
1. Imports matplotlib for plotting.
2. Plots the 'Open' prices for rows 1 to 19 with a green line and label "open".
3. Plots the 'Close' prices for the same range with a red line and label "close".
4. Adds a legend to distinguish the two lines.
5. Displays the plot.

This example demonstrates how to plot and compare specific columns over a selected range of data.
"""


import matplotlib.pyplot as plt

plt.plot(data["Open"][1:20], color="green", label="open")
plt.plot(data['Close'][1:20], color="red", label="close")
plt.legend()
plt.show()
```

```python
"""
This script prints subsets of the DataFrame `data` to show the beginning and the end of the dataset.

Operations:
1. Prints the first 3 rows using `data.head(3)`.
2. Prints the last 6 rows using `data.tail(6)`.

These methods help quickly inspect the structure and sample values from both ends of the DataFrame.
"""

print(data.head(3))
print(data.tail(6))
```

```python
"""
This script converts the 'Date' column of the DataFrame `data` to datetime format 
and then plots the 'Open' prices of BTC against these dates.

Steps:
1. Converts the 'Date' column to pandas datetime objects for proper time series handling.
2. Plots the 'Open' prices over time with grid lines.
3. Adds a title, x-axis label ('Date'), and y-axis label ('Price').
4. Saves the plot as "BTC_Open.png".
5. Displays the plot.

This is a common approach for visualizing time series financial data.
"""

data["Date"] = pd.to_datetime((data['Date']))

plt.plot(data['Date'], data['Open'])
plt.grid()

plt.title("BTC open prices")
plt.xlabel("Data")
plt.ylabel("Price")
plt.savefig("BTC_Open")
plt.show()
```

```python
"""
This script performs date conversion and data analysis on a BTC dataset with Persian calendar integration.

Steps:
1. Converts the 'Date' column to datetime objects.
2. Creates a new column 'Benefit' representing the daily profit (Close - Open).
3. Converts Gregorian dates to Persian (Jalali) dates and stores them in 'PersianDate'.
4. Extracts the Persian weekday number (1=Monday,...,7=Sunday) in 'PersianDay'.
5. Extracts the Persian calendar week number in 'PersianWeekOfYear'.
6. Prints the row(s) where the 'Benefit' is maximum, showing 'Benefit', 'PersianDate', and 'PersianDay'.
7. Calculates and prints the mean 'Benefit' grouped by 'PersianDay' to analyze average daily profits in the Persian week.

This script demonstrates combining datetime processing with cultural calendar conversion and grouping analysis.
"""

from persiantools.jdatetime import JalaliDate

data['Date'] = pd.to_datetime(data['Date'])
data['Benefit'] = data['Close'] - data['Open']
data['PersianDate'] = data['Date'].apply(lambda d: JalaliDate(d))
data["PersianDay"] = data['Date'].apply(lambda d: JalaliDate(d).isoweekday() )
data["PersianWeekOfYear"] = data['Date'].apply(lambda d: JalaliDate(d).week_of_year())


print(data[['Benefit', 'PersianDate', 'PersianDay']] [data['Benefit'] == data['Benefit'].max()])

print(data[["PersianDay", "Benefit"]].groupby('PersianDay').mean())
```
